#READING IMAGE USING OPENCV

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def showimage(img):
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#cvtColor() is used to convert an image from one colour space to another

#COLOR_BGR2RGB used to swap red and blue channels to use original channel sequence in images.
```

```python
import cv2

img = cv2.imread('bts oldies on Twitter.jfif')
```

```python
showimage(img)
```

#CONVERTING RGB TO GREYSCALE

```python
def twoGreyscale(img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

grey_scale_img = twoGreyscale(img)
plt.imshow(grey_scale_img, cmap ="gray")
```

```python
#printing intensity of greyscale image at location [34,54]
print(grey_scale_img[34][54])
```

```python
#printing intensity of RGB image
print(img[34][54])
```
