```python
import cv2
import matplotlib.pyplot as plt
```

##**PSEUDOCOLOURING**

```python
img=cv2.imread("/content/planet.jfif")
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap="gray")
```

```python
# Colourmap : JET
pseudo_image = cv2.applyColorMap(img, cv2.COLORMAP_JET)

# Display the pseudocolored image
plt.imshow(pseudo_image)
```

```python
#Colourmap: RAINBOW
pseudo_image2= cv2.applyColorMap(img, cv2.COLORMAP_HSV)

plt.imshow(pseudo_image2)
```
