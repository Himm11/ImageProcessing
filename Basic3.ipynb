##**IMAGE THRESHOLDING**

##1. Global Thresholding


```python
image=cv2.imread("/media/input_image.jpg")
```

```python
img= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap = "gray")     
```

## a. Binary thresholding

```python
#Setting threshold and Maximum value
t = 118
maxVal= 255

#Binary Thresholding
r, thresh1 =cv2.threshold(img, t, maxVal, cv2.THRESH_BINARY)

plt.imshow(thresh1, cmap="gray")
```

## b. Inverse BInary Thresholding

```python
r, thresh2= cv2.threshold(img , t, maxVal, cv2.THRESH_BINARY_INV)

plt.imshow(thresh2, cmap="gray")
```

##Truncate Thresholding

```python
r, thresh3= cv2.threshold(img , t, maxVal, cv2.THRESH_TRUNC)

plt.imshow(thresh3, cmap="gray")
```

##Threshold to Zero

```python
r, thresh4= cv2.threshold(img , t, maxVal, cv2.THRESH_TOZERO)

plt.imshow(thresh4, cmap="gray")
```

##Inverted Threshold to Zero

```python
r, thresh5= cv2.threshold(img , t, maxVal, cv2.THRESH_TOZERO_INV)

plt.imshow(thresh5, cmap="gray")
```

## 2. Otsu's method

```python
r, thresh6 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY +  cv2.THRESH_OTSU) 

plt.imshow(thresh6, cmap="gray")
```

```python
img3= cv2.imread("/content/20191208_230645.jpg")

img3= cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

plt.imshow(img3, cmap="gray")
```

```python
r, thresh6 = cv2.threshold(img3, 120, 255, cv2.THRESH_BINARY +  cv2.THRESH_OTSU) 

plt.imshow(thresh6, cmap="gray")
```

##3. Adaptive Thresholding

```python
#Adaptive mean (mean - constant as adaptive method)
#while using a smaller blocksize
thresh7= cv2.adaptiveThreshold(img3, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 21, 5)

plt.imshow(thresh7, cmap="gray")
```

```python
#while using a bigger blocksize
thresh8= cv2.adaptiveThreshold(img3, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 199, 5)

plt.imshow(thresh8, cmap="gray")
```

```python
#while using a larger mean value
thresh10= cv2.adaptiveThreshold(img3, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 199, 30)

plt.imshow(thresh10, cmap="gray")
```

```python
#Adaptive Guassian
thresh9=cv2.adaptiveThreshold(img3, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 199, 5)

plt.imshow(thresh9, cmap="gray")
```
